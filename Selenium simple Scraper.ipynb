```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC



def start_scraper_for_novel(beging, end, starting_url, text_file_name, content_tag, next_tag):
    
    options = Options()
    options.headless = True
    options.add_argument("--window-size=1920,1200")

    DRIVER_PATH = "/home/dev/Downloads/chromedriver"
    driver = webdriver.Chrome(executable_path=DRIVER_PATH)
    driver.get(starting_url)
    
    for x in range(beging, end):
        with open(f"{text_file_name}-{beging}-{end}.txt", "a") as myfile:             
            try:
                print(f"Starting to scrape chapter {x}")
                content = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CLASS_NAME, content_tag))
                ).text

                print(f"Managed to scrape the content of the chapter {x}")
                print(f"Wrinting in the file....")

                myfile.write(content)

                myfile.write("\n\n\n")
                myfile.write(f"End of {x} chapter")
                myfile.write("\n\n\n")
                myfile.write(f"New Chapter Begins: {x + 1}")
                myfile.write("\n\n\n")
                
                print("Searching for next chapter tag..")
                
                # driver.find_element(next_tag).send_keys(Keys.RETURN)
                driver.find_element_by_link_text(next_tag).click()
                print(f"Moving onto next chapter\n\n")


            except:
                print(f"stopped on chapter {x}, can't load page with url: {driver.current_url}")
                driver.quit()
                break

    print("End of task")
    driver.quit()
    
    
start_scraper_for_novel(1, 
                        40, 
                        "https://nyx-translation.com/2021/01/19/i-got-a-cheat-ability-in-a-different-world-and-become-extraordinary-even-in-the-real-world-vol-6-prologue-part-2/",
                       "i_got_cheat_ability",
                       "entry-content",
                       "Next >>")
```
